```python
from sklearn.model_selection import train_test_split
```

```python
#extracting data from excel file and saving it in a csv file.
exceldata = pd.read_excel('data.xlsx', index_col=0)
exceldata.to_csv('data.csv')
```

```python
csvdata = pd.read_csv('data.csv', index_col=0)
csvdata["inssurance"].replace({'yes': 1, 'no': 0}, inplace=True)
csvdata
```

```python
plt.scatter(x='age', y='inssurance', color = 'g', data=csvdata)
```

```python
x_train, x_test, y_train, y_test = train_test_split(csvdata[['age']], csvdata['inssurance'], test_size=0.2, random_state=42) # 80% training and 20% test
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train, y_train)
```

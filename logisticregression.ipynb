```python
import numpy as np 
import pandas as pd

```

```python
age = [25,32,47,52,23,40,33,36,28,50,22,27,35,45,26]
purchased = ['no','yes','yes','yes','no','yes','no','yes','no','yes','no','no','yes','yes','no']

dataset = pd.DataFrame({'age':age,'purchased':purchased})
dataset
```

```python
dataset.replace({'yes':1,'no':0},inplace=True)
dataset
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(dataset[['age']],dataset['purchased'],test_size=0.2,random_state=0)
```

```python
display(x_train, x_test, y_train, y_test)
```

```python
import matplotlib.pyplot as plt
plt.scatter(dataset['age'] , dataset['purchased'] , color = 'red')
plt.ylabel('purchased')
plt.xlabel('age')
plt.title('age vs purchased chart')
```

```python
# find accuracy, precision, recall, and f1 score
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
accuracy = accuracy_score(y_train, y_pred)
precision = precision_score(y_train, y_pred)
recall = recall_score(y_train, y_pred)
f1 = f1_score(y_train, y_pred)
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 score:", f1)
```

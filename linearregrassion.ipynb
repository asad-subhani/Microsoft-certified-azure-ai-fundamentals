```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
#given data
diameter = np.array([8, 10 ,12])
price = np.array([10, 13, 16])
```

```python
m, b = np.polyfit(diameter, price, 1) # y = mx + b 
```

```python
display(f"slope = {m}, intercept = {b}") # y = mx + b (m = slope, b)
```

```python
# ploting graph for my data
plt.scatter(diameter, price, color = 'b', label = 'data points') 
plt.plot(diameter, m * diameter + b, c = 'g', label = 'regression line') 
plt.xlabel ('diameter')
plt.ylabel ('price')
plt.title ('pizza price vs pizza diameter')
plt.show()
```

```python
display("Linear regression equation: y = {:.1f}x + {:.1f}".format(m, b))
```

```python
#predicting price for diameter = 13
price_predicted = m * diameter + b
```

```python
# updating the above grapg to show the predicted prices
testdiameters = np.array([7, 11, 13])

plt.scatter(diameter, price, color = 'b', label = 'data points') 
plt.plot(diameter, m * diameter + b, c = 'g', label = 'regression line') 
plt.scatter(testdiameters, m*testdiameters + b, color = 'r', label = 'predicted price')
plt.xlabel ('diameter')
plt.ylabel ('price')
plt.title ('pizza price vs pizza diameter')
plt.show()
```

```python
mae = mean_absolute_error(price, price_predicted)
mae
```

```python
col = [ 'Pclass', 'Sex']
df = df.drop(columns = col, axis=1)
df.info()
```

```python
df['Age'].fillna(df['Age'].mode()[0], inplace = True)
df['Cabin'].fillna(df['Cabin'].mode()[0], inplace = True)
df.info()
```

```python
df['Embarked'].unique()
```

```python
df['Embarked'].replace(['S', 'C', 'Q'], [0, 1, 2], inplace = True)
df['Embarked']
```

```python
df.head(10)
```

```python
min_fair = df['Fare'].min()
max_fair = df['Fare'].max()
df['normalized_fare'] = ((df['Fare'] - min_fair) / (max_fair - min_fair)).round(3)
df.head(10)
```

```python
age_group = {
    '0': (0, 20),
    '1': (20, 40),
    '2': (40, 60),
    '3': (60, 91),
    '4': (91, 200) 
}

def age_grouping(age):
    for key, (lower, upper) in age_group.items():
        if age >= lower and age <= upper:
            return key
    return -1

df['age_group'] = df['Age'].apply(age_grouping)
df.head(10)
```
